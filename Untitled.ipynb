```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

import scipy
import imageio

from numpy import pi
from numpy import sin
from numpy import zeros
from numpy import r_
from scipy import signal
from scipy import misc

%matplotlib inline
#pylab.rcParams["figure.figsize"] = (20, 7)
```

```python
WINDOW_SIZE = 16
```

```python
im = imageio.imread("a.png")
f = plt.figure()
plt.imshow(im[:, :, 0], cmap="gray")
```

```python
def dct2(a):
    return scipy.fftpack.dct(scipy.fftpack.dct(a, axis=0, norm="ortho"), axis=1, norm="ortho")

def idct(a):
    return scipy.fftpack.idct(scipy.fftpack.idct(a, axis=0, norm="ortho"), axis=1, norm="ortho")
```

```python
imsize = im.shape
dct = np.zeros((imsize[0], imsize[1]))

for i in r_[:imsize[0]:WINDOW_SIZE]:
    for j in r_[:imsize[1]:WINDOW_SIZE]:
        dct[i : (i + WINDOW_SIZE), j : (j + WINDOW_SIZE)] = dct2(im[i : (i + WINDOW_SIZE), j : (j + WINDOW_SIZE), 0])
```

```python
pos = 0

plt.figure()
plt.imshow(im[pos : pos + WINDOW_SIZE, pos : pos + WINDOW_SIZE, 0], cmap="gray")
plt.title("An 16x16 Image block")

plt.figure()
plt.imshow(dct[pos : pos + WINDOW_SIZE, pos : pos + WINDOW_SIZE], cmap="gray", vmax = np.max(dct) * 0.01, vmin=0, extent=[0, pi, pi, 0])
plt.title("An 16x16 DCT block")
```

```python
# DISPLAY ALL BLOCKS
plt.figure()
plt.imshow(dct[:, :], cmap="gray", vmax=np.max(dct) * 0.01, vmin=0)
plt.title("16x16 entire DCT image")
```

```python
# Threshold
thresh = 0.011

dct_thresh = dct * (abs(dct) > (thresh * np.max(dct)))

plt.figure()
plt.imshow(dct_thresh[:, :], cmap="gray", vmax=np.max(dct) * 0.01, vmin=0)
plt.title("Thresholded 8x8 DCT image")

percent_nonzeros = np.sum(dct_thresh != 0.0) / (imsize[0] * imsize[1] * 1.0)

print("Keeping only %f%% of the DCT coefficients" % (percent_nonzeros* 100.0))
```

```python
unique, counts = np.unique(dct_thresh, return_counts=True)
deneme = dict(zip(unique, counts))
print(dct_thresh.shape, len(unique))
```

```python
for a, b in deneme.items():
    if b > 15:
        dct_thresh[dct_thresh == a] = 255
        print(a, b)
        print("Benzerlik bulundu")
```
